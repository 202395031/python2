#판다스의 DataFrame() 함수로 데이터 프레임을 만들기.

*pd.DataFrame(값,columns=열 이름 리스트)*

```python
# 리스트를 만들어 데이터로 한 행씩 지정.
list1 = list([
    ['홍길동', '남자', '20', '180'],
    ['김길동', '남자', '21', '177'],
    ['박길동', '여자', '20', '160']
])

# 열 이름 리스트 생성.
col_names = ['이름', '성별', '나이', '키']

pd.DataFrame(list1,columns=col_names)   # 가로안에는 순서대로 한 행에 들어갈 값과, 열의 이름이다.
```

###**조건식에 isin()함수, and기호(&), or 기호(|), str.contains()함수를 사용하여 더 복잡한 조건으로 조회할 수도 있음**

```python
# 나이가 21세 또는 23인 사람을 조회.
df[df['나이'].isin([21,23])]
```

```python
# 성별이 '여자' 이면서 키가 160보다 큰 사람을 조회.
# and => &
df[(df['성별'] == '여자') & (df['키'] >= 160)]
```

```python
# 나이가 28세 이상이거나 성별이 '남자' 인 사람을 조회.
# or => |
df[(df['나이'] >= 28) | (df['성별'] == '남자')]
```

```python
# 이름에 '순'이 들어간 사람을 조회.
df[df['이름'].str.contains('순')]
```

```python
# 나이와 키에 대한 통계(int값의 통계) = > 숫자형 데이터만 통계를 구해준다.
df.describe()
```

##**성별을 기준으로 조회하여 키의 평균과 표준편차를 계산.**

```python
df.describe()

# 평균 => mean(), 표준편차 => std()
```

```python
# 성별의 평균 키
# 성별을 기준으로 그릅화 하여 그룹별 키의 평균을 계산
# as_index=False => 데이터프레임으로 출력 됨.
mean_by_gender = df.groupby(by=['성별'], as_index=False)['키'].mean()

# 열 이름 바꾸기
mean_by_gender.rename(columns={'키':'평균키'}, inplace=True)

mean_by_gender
```

```python
# 성별의 키의 표준 편차
std_by_gender = df.groupby(by=['성별'], as_index=False)['키'].std()
std_by_gender.rename(columns={'키':'키의 표준편차'}, inplace=True)

std_by_gender
```

```python
# 두 데이터 프레임을 합치기
new_df = pd.merge(mean_by_gender, std_by_gender)

new_df
```

#**시간표 분석**

```python
# 열 이름과 시간표 데이터를 리스트로 저장.
col_names = ['과목번호', '과목명', '강의실', '시간수']
list1 = list([['C1', '파이썬', '303호', 3],
              ['C2', '국어', '201호', 3],
              ['C3', '영어', '203호', 3],
              ['C4', '한국사', '201호', 2],
              ['C5', '과학', '401호', 3],
              ['C6', '사회', '201호', 2]])
```

```python
# 시간표 데이터를 데이터프레임 객체 df로 변환하여 CSV 파일로 저장.
import pandas as pd

# 데이터 프레임으로 변환
timetable_df = pd.DataFrame(list1, columns = col_names)

timetable_df.to_csv('./timetable.csv', header=True, index=False, encoding='utf-8')
```

```python
# csv 파일 읽어와서 출력하기
timetable_df = pd.read_csv('./timetable.csv', sep=',')

timetable_df
```

```python
# 열 이름 '담담 교사' 추가
# 값 => '이순신', '김유신', '강감찬', '신사임당', '정약용', '허준'
timetable_df['담담교사'] = ['이순신', '김유신', '강감찬', '신사임당', '정약용', '허준']
timetable_df
```

```python
# 강의실별 최대시간 구하기.
max_hour = timetable_df.groupby(by=['강의실'], as_index=False)['시간수'].max()
max_hour.rename(columns={'시간수':'최대시간'}, inplace=True)
max_hour
```
